```python
import matplotlib.pyplot as plt
import numpy as np

# Logged RL rewards (example run)
episodes = np.array([0, 10, 20, 30, 40])
total_reward = np.array([-23.98, 4.98, -3.44, 6.19, -1.49])

plt.figure(figsize=(6,4))
plt.plot(episodes, total_reward, marker='o')

plt.xlabel("Training Episode")
plt.ylabel("Total Episode Reward")
plt.title("MAPPO + SEOBM Training Signal")
plt.grid(True)
plt.tight_layout()
plt.show()
```
