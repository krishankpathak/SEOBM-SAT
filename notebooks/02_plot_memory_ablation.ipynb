```python
import matplotlib.pyplot as plt
import numpy as np

# Representative ablation results
memory_window = np.array([1, 5, 10])
mse = np.array([0.062, 0.044, 0.036])

plt.figure(figsize=(5,4))
plt.plot(memory_window, mse, marker='o')

plt.xlabel("Memory Window Size (K)")
plt.ylabel("MSE (normalized)")
plt.title("Effect of SEOBM Memory Depth")
plt.grid(True)
plt.tight_layout()
plt.show()
```
