```python
import matplotlib.pyplot as plt
import numpy as np

# Logged results from delayed prediction run
epochs = np.array([0, 10, 20, 30, 40])
baseline_mse = np.array([0.1122, 0.0700, 0.0435, 0.0301, 0.0244])
seobm_mse = np.array([0.0644, 0.0567, 0.0523, 0.0461, 0.0357])

plt.figure(figsize=(6,4))
plt.plot(epochs, baseline_mse, marker='o', label="Baseline (no memory)")
plt.plot(epochs, seobm_mse, marker='s', label="SEOBM (temporal memory)")

plt.xlabel("Training Epoch")
plt.ylabel("MSE (normalized)")
plt.title("Delayed Prediction Performance")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
